```python
import pandas as pd
import matplotlib.pyplot as plt
crime = pd.read_csv('Crime_Data_from_2020_to_Present_20250609.csv')
```

```python
df = crime.drop(labels= ['Mocodes', 'Weapon Used Cd', 'Weapon Desc', 'Crm Cd 2', 'Crm Cd 3', 'Crm Cd 4', 'Cross Street'], axis=1)
df
```

```python
df.sort_values(by = 'Date Rptd')
```

```python
df['Date Rptd'] = pd.to_datetime(df['Date Rptd'], format='%m/%d/%Y %I:%M:%S %p', errors='coerce')
df['DATE OCC'] = pd.to_datetime(df['DATE OCC'], format='%m/%d/%Y %I:%M:%S %p', errors='coerce')
```

```python
df.info()
```

```python
df.describe(include = ['float64', 'int64'])
```

```python
df['Vict Age'].value_counts()
```

```python
df_age = df[df['Vict Age'] > 0]
```

```python
df['year'] = df['DATE OCC'].dt.year
df['month'] = df['DATE OCC'].dt.month
```

```python
df
```

Feature Understanding

```python
df['month'].value_counts().sort_index().plot()
```

```python
df['year'].value_counts().sort_index().plot(kind='bar')
```

```python
df = df[~((df['month'] == 12) & (df['year'] == 2025))]
```

```python
df.sort_values(by = 'DATE OCC')
```

```python
df.groupby(['year', 'month']).size().plot()
```

```python
plt.hist(df_age['Vict Age'], edgecolor = 'black', bins=15)
```

```python
crime_count_per_day = df.groupby(['DATE OCC', 'Crm Cd Desc']).size()
```

```python
top_20_crimes = crime_count_per_day.groupby('Crm Cd Desc').mean().reset_index(name='average daily count').sort_values(by='average daily count', ascending=False).head(20)
```

```python
top_20_crimes = top_20_crimes.rename(columns={'Crm Cd Desc': 'Crime description'})
```

```python
top_20_crimes_sorted = top_20_crimes.sort_values(by = 'average daily count', ascending=True)
```

```python
plt.barh(top_20_crimes_sorted['Crime description'], top_20_crimes_sorted['average daily count'])
```

```python
plt.figure(figsize=(12,8))

#Plot 1 - Line Chart
plt.subplot(2,2,1)
df.groupby(['year', 'month']).size().plot().set_xlabel('Year - Month', fontsize = 10)
plt.title('Crime Over Time')
#Plot 2 - Histogram
plt.subplot(2,2,2)
plt.hist(df_age['Vict Age'], edgecolor = 'black', bins=15)
plt.tight_layout()
plt.title('Distribution of Victim Age')

#Plot 3 - Horizontal Bar
plt.figure()
plt.barh(top_20_crimes_sorted['Crime description'], top_20_crimes_sorted['average daily count'])
plt.title('Top 20 Crimes')
```
